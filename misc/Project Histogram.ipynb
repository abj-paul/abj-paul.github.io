```python
import matplotlib.pyplot as plt

# Sample data
projects = {
    "Etherium/Solana Contract": 7,
    "NLP": 15,
    "Computer Vision": 3,
    "Web Design": 10
}
colors = ['purple', 'lightblue', 'green', 'orange']  # List of colors for each bar

plt.barh(range(len(projects.values())), projects.values(), color=colors)

# Adding labels and title
plt.xlabel('Project Count')
plt.ylabel('Techstack')
plt.title('Skills')

index = 0
for technology, project_count in projects.items():
    plt.text(project_count, index, technology, ha='left', va='center')
    index +=1
plt.yticks([])

# Remove boundary around plot
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(True)
plt.gca().spines['left'].set_visible(False)
plt.show()
```
